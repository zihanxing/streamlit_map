```python
import pandas as pd
import numpy as np
```

```python
df_disater = pd.read_csv('Total_Disasters_2024.csv')
df_deaths = pd.read_csv('Number_of_Deaths_2024.csv')
df_financal = pd.read_csv('Total_Financial_Damage_2024.csv')
df_injuries = pd.read_csv('Number_of_Injuries_2024.csv')
```

```python
df_2019 = pd.read_csv('data_since_2019.csv')
```

```python
# Select only the columns we need 'State Code', 'Predicted Total Disasters for 2024'
# df_disater = df_disater[['State Code', 'Predicted Total Disasters for 2024']]
```

```python
df_disater = df_disater.rename(columns={'Predicted Number of Total Disasters for 2024': 'Total Disasters'})
df_deaths = df_deaths.rename(columns={'Predicted Total Deaths for 2024':'Total Deaths'})
df_financal = df_financal.rename(columns={'Predicted Number of Financial Damage for 2024':'Total Damage (\'000 US$)'})
df_injuries = df_injuries.rename(columns={'Predicted Number of Injuries for 2024':'No. Injured'})
```

```python
df_all = df_disater.merge(df_deaths, on='State Code', how='outer').merge(df_financal, on='State Code', how='outer').merge(df_injuries, on='State Code', how='left')
```

```python
# set all nagative values in df_all to 0, except for "State Code" column
df_all[df_all.columns.difference(['State Code'])] = df_all[df_all.columns.difference(['State Code'])].clip(lower=0)
```

```python
# Add State Name column, if State Code is not in the mapping, then the State Name will be Unknown
df_all['State Name'] = df_all['State Code'].map(state_name_mapping).fillna('Unknown')
```

```python
df_all['Year'] = 2024
```

```python
'''
low-risk is < 17.5 deaths
Medium-risk is >17.5 and <77 deaths
High-risk is >=77 deaths
'''

# Add Risk Level column
df_all['Risk Level'] = np.where(df_all['Total Deaths'] < 17.5, 1, np.where(df_all['Total Deaths'] < 77, 2, 3))
```
